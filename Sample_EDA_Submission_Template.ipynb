## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1. Standardize Gender
male_variants = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man", "msle", "mail", "malr", "cis man", "cis male"]
female_variants = ["cis female", "f", "female", "woman",  "femake", "female ", "cis-female/femme", "female (cis)", "femail"]
df['Gender'] = df['Gender'].str.lower()
df['Gender'] = df['Gender'].apply(lambda x: "Male" if x in male_variants else ("Female" if x in female_variants else "Other"))

# 2. Fix Age
df = df[(df['Age'] >= 18) & (df['Age'] <= 80)].copy()

# 3. Fill nulls
df['self_employed'] = df['self_employed'].fillna(df['self_employed'].mode()[0])
df['work_interfere'] = df['work_interfere'].fillna(df['work_interfere'].mode()[0])

# 4. Drop columns
df.drop(columns=['Timestamp', 'State', 'comments'], inplace=True, errors='ignore')
df.reset_index(drop=True, inplace=True)
````

### What all manipulations have you done and insights you found?

### What all manipulations have you done and insights you found?

First, I mapped all the free-text variants of gender down to "Male", "Female", and "Other" to enable structured querying and to resolve dirty inputs. Second, I filtered out nonsensical ages by ensuring all respondents were between 18 and 80 years old, which correctly removed extreme outliers and negative values. Third, I imputed missing categorical values in the 'self_employed' and 'work_interfere' features by using their respective mathematical modes, ensuring no rows are needlessly dropped while capturing the most represented tendency. Finally, I dropped 'Timestamp', 'State' (which inherently excludes the majority non-US cohort or is too sparse to matter globally), and 'comments' (which had an overwhelming rate of null textual inputs) to strip out noise ahead of the visualization phase. The insight here is the dataset now accurately reflects clean analytical cohorts without disruptive NaN leakage.

#### Chart - 14 - Correlation Heatmap

````python
# Chart - 14 - Correlation Heatmap visualization code
# Encode binary/ordinal columns first, then heatmap
df_encoded = df.copy()
cols_to_encode = ['Gender', 'family_history', 'treatment', 'remote_work', 'tech_company', 'benefits', 'work_interfere']
for col in cols_to_encode:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

plt.figure(figsize=(10,5))
corr = df_encoded[['Age'] + cols_to_encode].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap reveals the Pearson linear coefficients between computationally encoded features in one sweeping glance, quickly eliminating variables devoid of predictive value.

##### 2. What is/are the insight(s) found from the chart?

The strongest positive linear driver for 'treatment' is 'family_history'. Conversely, basic demographic trackers like 'Age' or 'remote_work' show basically zero correlative traction.
